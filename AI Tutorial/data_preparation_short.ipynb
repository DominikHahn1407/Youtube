```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

Create DF

```python
df = pd.DataFrame({"age": [22, 35, 58], "income": [30000, 60000, 90000]})
df.head()
```

Create Features

```python
df["income_per_age"] = round(df["income"] / df["age"], 2)
df["age_group"] = pd.cut(df["age"], bins=[0, 30, 50, 100], 
                         labels=["young", "adult", "senior"])
df.head()
```

Categorical Encoding

```python
df = pd.get_dummies(df, columns=["age_group"]) 
df.head()
```

Scale Features

```python
scaler = StandardScaler()
df[["age", "income", "income_per_age"]] = scaler.fit_transform(df[["age", "income", "income_per_age"]])
df.head()
```
