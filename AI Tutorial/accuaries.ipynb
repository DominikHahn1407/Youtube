```python
y_true = [0, 0, 0, 0, 0, 0, 1, 1, 2, 2]
y_pred = [0, 0, 0, 0, 0, 1, 1, 0, 0, 2]

classes = sorted(set(y_true))

print("True labels:     ", y_true)
print("Predicted labels:", y_pred)
print()
```

## Overall Accuracy

```python
correct = 0

for true, pred in zip(y_true, y_pred):
    if true == pred:
        correct += 1

overall_accuracy = correct / len(y_true)
print(f"Overall accuracy: {overall_accuracy:.2f}")
```

## Balanced Accuracy

```python
class_accuracies = []

for cls in classes:
    total_in_class = 0
    correct_in_class = 0

    for true, pred in zip(y_true, y_pred):
        if true == cls:
            total_in_class += 1

            if pred == cls:
                correct_in_class += 1

    class_accuracy = correct_in_class / total_in_class
    class_accuracies.append(class_accuracy)

balanced_accuracy = sum(class_accuracies) / len(class_accuracies)
print(f"Balanced accuracy: {balanced_accuracy:.2f}")
```
