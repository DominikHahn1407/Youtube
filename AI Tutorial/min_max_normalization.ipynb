```python
values = [1, 2, 3, 4, 5]

max_val = max(values)
min_val = min(values)

normalized = [(x - min_val) / (max_val - min_val) for x in values]
normalized
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sk_normalized = scaler.fit_transform([[x] for x in values])
sk_normalized
```
